```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt 
#Reading data as a dataframe and storing it in df variable 
df=pd.read_csv("netflix_data.csv")
df 
#we have 9 string columns and 2 numerical columns 
```

```python
#check for duplicate rows
duplicate_rows=df.duplicated().sum()
print("The number of duplicate rows are:", duplicate_rows)
```

```python
#counting number of null values for each column
null_values=df.isnull().sum()
print("Missing Values are:\n", null_values)
#we have null values in director, cast, country and date_added columns 
```

```python
#Handling missing values in each column with NaN
df['director']=df['director'].fillna("NaN", inplace=True)
df['cast']=df['cast'].fillna("NaN", inplace=True)
df['country']=df['country'].fillna("NaN", inplace=True)
df['date_added']=df['date_added'].fillna("NaN", inplace=True)
```

```python
df.isnull().sum()
```

```python
#Filter the data for movies between 1990-1999
netflix_subset_movies=df[(df['type']=='Movie') & (df['release_year'].between(1990,1999))]
netflix_subset_movies
```

```python
# Calculate the frequency of each movie
duration_count=netflix_subset_movies['duration'].value_counts()
# Most frequent Movie duration
duration=duration_count.idxmax()
print("The most frequent movie duration in the 1990s is: ",duration)
```

```python
#visualize the distribution of duration for movie released in the 1990s
import seaborn as sns
#plot the distribution of movie durations
plt.figure(figsize=(10,6))
sns.histplot(data=netflix_subset_movies, x='duration', bins=20, kde=True)
plt.title('Distribution of Movie Durations(1990s)')
plt.xlabel('Duration(minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
#count the number of short movies in the 1990s
short_movies=netflix_subset_movies[netflix_subset_movies['duration']<90]
short_movie_count=short_movies.shape[0]
print("The number of short movies in the 1990s is:", short_movie_count)
```

```python
#For 'House of Cards', what is the show is and who is the director of this show
df[df['title'].isin(['House of Cards'])]
#isin(): To show all records of a particular item in any column
```

```python
#In which year highest number of the Tv shows & Movies were released? 
plt.figure(figsize=(14,6))
df['release_year']=pd.to_numeric(df['release_year'],errors='coerce')
counts=df['release_year'].value_counts().sort_index()
counts.plot(kind='line', marker='o')
plt.title('Netflix Releases Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#How many movies and Tv shows are in the dataset
df.groupby('type').genre.count()
```

```python
df['type'].value_counts().plot(kind='bar',color=['salmon','skyblue'])
plt.title('count of Movies and Tv shows')
plt.xlabel('Type')
plt.ylabel('Number of Titles')
plt.show()
```

```python
#Top 10 Genres of Movies Released in 2000
movies_2000 = df[(df['type'] == 'Movie') & (df['release_year'] == 2000)]
movies_2000
```

```python
genres_2000=movies_2000['genre'].str.split(', ').explode()
top_genres=genres_2000.value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='Set2')
plt.title('Top 10 Genres of Movies Released in 2000')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
#show only the Titles of all Tv shows that were released in India only 
titles = df[(df['type']=='TV Show') & (df['country']=='India')]['title'].tolist()
for title in titles:
    print(title)
```

```python
print("Number of TV Shows released in India:", len(titles))
```

```python
#Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom".
filtered_df=df[((df['type']=="Movie")& (df['genre']=='Comedies')) | (df['country']=='United Kingdom')]
filtered_df
```

```python
# Count genres in filtered data
genre_counts = filtered_df['genre'].value_counts().head(10)  
plt.figure(figsize=(10, 5))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='flare')
plt.title('Top Genres (Movies that are Comedies or from UK)')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
df['duration']=df['duration'].astype(str)
print(df.duration.dtype)
```

```python
df['duration'] = df['duration'].astype(str)
split_duration = df['duration'].str.split(' ', n=1, expand=True)
print(split_duration.head())
df['Minutes'] = split_duration[0]  
df['Unit'] = split_duration[1] if split_duration.shape[1] > 1 else None
```

```python
import seaborn as sns
#Set style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
#Top 10 genres by count
top_genres = df['genre'].value_counts().head(10)
plt.figure()
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
#Movies vs TV Shows over the years
content_trend = df.groupby(['release_year', 'type']).size().reset_index(name='count')
plt.figure()
sns.lineplot(data=content_trend, x='release_year', y='count', hue='type', marker='o')
plt.title("Movies vs TV Shows Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Releases")
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Countries by Content Count
top_countries = df['country'].value_counts().head(10)
plt.figure()
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
